# Cancer Detection AI

This notebook train a AI to detect cancer which is malignant or benign.

Data is downloaded from: https://gist.github.com/KhanradCoder/35a6beea49e5b9ba62797e595a9626c0.

```python
# import required liraries
import pandas as pd
import os

from sklearn.model_selection import train_test_split

import tensorflow as tf
```

```python
SEED: int = 30024
```

```python
# read 'cancer.csv', please ensure the working directory is correct
dataset = pd.read_csv('./Cancer Detection AI/cancer.csv')
```

```python
dataset.tail(5)
```

```python
dataset_nolabel = dataset.drop(dataset.columns[0], axis=1)
dataset_nolabel.tail(5)
```

```python
dataset_label = dataset[dataset.columns[0]]
dataset_label.tail(5)
```

```python
# assign 20% data as testing set
test_size = 0.2

# split the data into training set and testing set
dataset_train, dataset_test, label_train, label_test = train_test_split(dataset_nolabel, 
                                                                        dataset_label, 
                                                                        test_size=test_size, 
                                                                        random_state=SEED)
```

```python
# construct a sequential model
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(256, input_shape=dataset_train.shape[1:], activation = 'sigmoid'),
    tf.keras.layers.Dense(256, activation='sigmoid'),
    tf.keras.layers.Dense(1, activation='sigmoid')
], name='CancerDetector')

# display summary info
model.summary()
```

```python
# compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# define a callback function to stop the epoch while reach the expectations
class myCallbacks(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if logs.get('accuracy') is not None and logs.get('accuracy') > 0.975 and epoch > 500:
            self.model.stop_training = True
            
# initilizing callbacks
callbacks = myCallbacks()
```

```python
# fit the dataset
model.fit(dataset_train, label_train, epochs=1000, callbacks=[callbacks])
```

```python
# evaluate the model performance
model.evaluate(dataset_test, label_test)
```
